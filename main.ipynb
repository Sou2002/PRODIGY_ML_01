### **Importing necessary libraries**

```python
import pandas as pd
pd.options.mode.chained_assignment = None # To eliminate warnings due to chained operations
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

### **Importing Train & Test Data**

```python
train_df = pd.read_csv("datasets/train.csv")
train_df.head()
```

```python
train_df.columns
```

```python
train_df.shape
```

```python
test_df = pd.read_csv("datasets/test.csv")
test_df.head()
```

```python
test_df.shape
```

### **Extracting Features (Based on Task Criteria)**

```python
train_df = train_df[["LotArea", "BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath", "BedroomAbvGr", "SalePrice"]]
train_df.head()
```

```python
train_df.info() #No Null values present
```

```python
train_df.describe()
```

### **Preprocessing**

```python
train_df["Bathroom"] = train_df.BsmtFullBath + train_df.BsmtHalfBath + train_df.FullBath + train_df.HalfBath
train_df.drop(["BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath"], axis = 1, inplace = True)
train_df.rename(columns = {"BedroomAbvGr": "Bedroom", "LotArea": "Area"}, inplace = True)
train_df
```

### **Outlier Detection**

```python
train_df.Area.skew()
```

```python
train_df.Area.plot(kind = 'box')
```

```python
train_df.Bedroom.skew()
```

```python
train_df.Bedroom.plot(kind = 'box')
```

```python
train_df.Bathroom.skew()
```

```python
train_df.Bathroom.plot(kind = 'box')
```

### **Outlier Removal**

```python
train_df = train_df[(train_df.Area <= 19000) 
                    & (train_df.Bedroom >= 1) 
                    & (train_df.Bedroom <= 4) 
                    & (train_df.Bathroom <= 4)]

train_df.skew()
```

### **Train-Validation-Test Split**

```python
X = train_df.drop("SalePrice", axis = 1)
X.head()
```

```python
y = train_df["SalePrice"]
y
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_test = test_df[["LotArea", "BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath", "BedroomAbvGr"]]
X_test.head()
```

```python
X_test.info()
```

```python
X_test[X_test.BsmtFullBath.isna()] # Checking Null values in Test Dataset
```

```python
X_test.dropna(inplace = True) # Removing Null values from Test Dataset
```

```python
X_test["Bathroom"] = (X_test.BsmtFullBath + X_test.BsmtHalfBath + X_test.FullBath + X_test.HalfBath).astype(int)
X_test.drop(["BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath"], axis = 1, inplace = True)
X_test.rename(columns = {"BedroomAbvGr": "Bedroom", "LotArea": "Area"}, inplace = True)
X_test
```

### **Scaling**

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_val_scaled = scaler.transform(X_val)
```

```python
X_test_scaled = scaler.transform(X_test)
```

### **Model Training**

```python
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_train_scaled, y_train) # Score is low beacuse we have dropped too many features
```

### **Model Testing**

```python
model.score(X_val_scaled, y_val)
```

### **Predicting using Model**

```python
y_pred = model.predict(X_test_scaled)
y_pred
```
